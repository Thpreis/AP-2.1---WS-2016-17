```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

##Berechnung der Beugungsbilder

#Beugungsfunktion Spalt
def beugung_spalt(x):
    return np.sinc(x)**2

#Beugungsfunktion Doppelspalt
def beugung_doppelspalt(x):
    return np.sinc(x)**2*np.cos(np.pi*v*x)**2

#Anzahl der Nebenmaxima, die dargestellt werden sollen:
a=1
#Verhältnis Spaltabstand zu Spaltbreite
#v=2
v=1.92

#Abszissenwerte
x = np.linspace(-(a+1), a+1, 200)

#Plot der Beugung
plt.plot(x, beugung_spalt(x), label="Spalt")
plt.plot(x, beugung_doppelspalt(x), label="Doppelspalt")
plt.xlabel("$\pi x$")
plt.ylabel("Intensität [b.E.]")
plt.title("Beugung am Einfach- und Doppelspalt")
plt.ylim(0, 1.1)
plt.legend()
#plt.savefig("V233/Plots/Begung_spalt_doppelspalt1.pdf", format="pdf")

#Drucke diesen Plot
```

```python
##Berechnung der modifizierten Spaltbilder

#Integrand
def spalt(k):
    return d/np.pi*np.sin(k*d/2)/(k*d/2)*np.cos(y*k)

#Spaltbreite auf 1
d=1

#Integrationsbereich
n=3

#y-Werte
Y=np.linspace(-1,1,200)*d

#Berechnung des resultierenden Objektbildes
from scipy.integrate import quad
f_modifiziert=[]
for i in range(len(Y)):
    y=Y[i]
    result, error = quad(spalt, 0, 2*np.pi*n/d)
    f_modifiziert.append(result**2)
    
#Kurve auf 1 normieren:
f_modifiziert = f_modifiziert/np.max(f_modifiziert)

#Plot
plt.plot(Y, f_modifiziert)
plt.xlabel("y / d")
plt.ylabel("Intensität [b.E]")
plt.title("Modifiziertes Objektbild Spalt")
plt.ylim(0,1.1)
plt.xlim(-1,1)
plt.savefig("V233/Plots/spalt3.pdf", format="pdf")



##Analog zu vorher ...
```

```python
#Doppelspalt
def doppelspalt(k):
    return d/np.pi*np.cos(k*g/2)*np.sin(k*d/2)/(k*d/2)*np.cos(y*k)

#Spaltabstand in Einheiten der Spaltbreite:
g=2*d
n=3

#y-Werte
Y=np.linspace(-3,3,800)*d

#Berechnung des resultierenden Objektbildes
from scipy.integrate import quad
f_modifiziert=[]
for i in range(len(Y)):
    y=Y[i]
    result, error = quad(doppelspalt, 0, 2*np.pi*n/d)
    f_modifiziert.append(result**2)
    
#Kurve auf 1 normieren:
f_modifiziert = f_modifiziert/np.max(f_modifiziert)

#Plot
plt.plot(Y, f_modifiziert)
plt.xlabel("y / d")
plt.ylabel("Intensität [b.E]")
plt.title("Modifiziertes Objektbild Spalt")
plt.ylim(0,1.1)
plt.xlim(-2,2)
plt.savefig("V233/Plots/Doppelspaltspalt.pdf", format="pdf")
```
