```python
#all things that need to be imported and definitions
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def comma_to_float(valstr):
    return float(valstr.decode("utf-8"). replace(',', '.'))
```

```python
#IX.3Zuordnung der fehundenen Linien zu Serien
##IX.3.1 Erwartete Linien fr die 1. Nebenserie

hc = 1.2398e3#-nmeV speed of light times Plackkonstant

lam_31 = 819.2 #reference lambda in nm
del_lam_31 = 2

m_lam_31 = 3#Ordnung

E_ry_31 = -13.605#eV see formula 7

E_3p_31 = E_ry_31/(m_lam_31**2) - hc/lam_31#using formula 7
del_E_3p_31 = hc/(lam_31**2)*del_lam_31
print("E_3p = ", E_3p_31, "+-", del_E_3p_31)

m_31 = np.arange(3,13,1)#Ordnung der Linine erster Nebenserie

lam_ex_31 = hc/(E_ry_31/m_31**2 - E_3p_31)#in nm expected wavelenghts of first order
del_lam_ex_31 = hc/(E_ry_31/m_31**2 - E_3p_31)**2*del_E_3p_31

for i in range(len(m_31)):
    #l=1.2398E3/(-13.605/m**2-E_3p)
    print('m={m:2d} lambda_ex_31 ={l:6.2f} +- del_lam_ex_31 = {del_lam:6.2f}'.
          format(m= m_31[i],l= lam_ex_31[i], del_lam = del_lam_ex_31[i] ))
```

```python
from scipy.optimize import curve_fit
from scipy.stats import chi2

##First series
def comma_to_float(valstr):
    return float(valstr. decode("utf-8"). replace(',', '.'))
#import experimentally identified wavelenghts and their order
#lam_4_1 in nm
m_4_1, lam_4_1, del_lam_4_1 = np. loadtxt('IX4_Wellenlaengen_1_Nebenserie.txt', skiprows=0,
    converters= {0:comma_to_float,1:comma_to_float},
    comments= '>', unpack= True)

#Function to calculate E_ry_4, E_3p_4, delta_d_4 for first series
def fit_func(m,E_Ry,E_3p,D_d):
    return 1.2398E3/(E_Ry/(m- D_d)**2-E_3p)

#calculate fitparameter first series
para= [-13.6, -3, -0.02]
popt1, pcov1= curve_fit(fit_func, m_4_1, lam_4_1,
        sigma= del_lam_4_1 ,p0= para)

print("E_Ry1_4 =",popt1[0],", Standardfehler=", np. sqrt(pcov1[0][0]), "eV")
print("E_3p1_4 =",popt1[1],", Standardfehler=", np. sqrt(pcov1[1][1]), "eV")
print("Delta_d=",popt1[2],", Standardfehler=", np. sqrt(pcov1[2][2]), "no unit?")

#calculate Chiquadrat and Chiquadratreduziert
chi2_1= np. sum((fit_func(m_4_1,*popt1)- lam_4_1)**2/del_lam_4_1**2)#maybe we need to change alle *popt1 into *popt
dof_1= len(m_4_1)-3#dof:degrees of freedom
chi2_red_1= chi2_1/dof_1
print("chi2_1 =", chi2_1)
print("chi2_red_1=",chi2_red_1)

#calculate "Fitwahrscheinlichkeit, Wahrscheinlichkeit, dass bei einer erneuten Messung ein chi_2 erhalten, der groeer oder gleich dem hier bestimmten ist.
prob= round(1- chi2.cdf(chi2_1,dof_1),2)*100#maybe need np.round?
print("Wahrscheinlichkeit:", prob,"%")

#plot measured wavelenghts of first series and include fitcurve
plt. errorbar(m_4_1, lam_4_1, del_lam_4_1, fmt= ".")
plt. xlabel('Quantenzahl')
plt. ylabel('Wellenlaenge / nm')
plt. title('1. Nebenserie des Na-Atoms')
x= np. linspace(2.8, 12.2, 100)
plt. plot(x, fit_func(x,*popt1))
plt. savefig("Plots/1_Nebenserie_4.pdf", format= "pdf")

```
