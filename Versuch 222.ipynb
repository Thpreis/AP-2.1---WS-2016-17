```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline

#variables
freq = np.array([4.71, 4.31, 4.11, 3.54])
wirk_th = np.array([0.07, 0.0671, 0.0655, 0.0597])
error_th = np.array([0.0029, 0.0027, 0.0026, 0.0026])
wirk_eff = np.array([0.0111, 0.0177, 0.0249, 0.0286])
error_eff = np.array([0.0009, 0.0010, 0.0012, 0.0013])

#both plots
plt.errorbar(freq, wirk_th, yerr=error_th, label = '$\eta_{th}$')
plt.errorbar(freq, wirk_eff, yerr=error_eff, label = '$\eta_{eff}$', color="red")

#title
plt.title('Wirkungsgrad in Abh√§ngigkeit der Frequenz')
plt.xlim(3.4, 5.2)
plt.ylim(0.003, 0.08)
#label
plt.xlabel(r'Frequenz $ f =[\mathrm{1/s}]$')
plt.ylabel(r'Wirkungsgrad $ \eta $')

#position of legend
plt.legend(loc='upper right')

plt.savefig('wirk_freq.pdf')
```
